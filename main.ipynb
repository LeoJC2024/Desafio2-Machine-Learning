```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import classification_report, accuracy_score


import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

# =====================================================================
# ETAPA 1: FONTE DE DADOS E OBJETIVO
# =====================================================================
print("\n" + "="*60)
print(" ETAPA 1: FONTE DE DADOS E OBJETIVO ".center(60, "="))
print("="*60)

url = "https://raw.githubusercontent.com/stedy/Machine-Learning-with-R-datasets/master/usedcars.csv"
dados = pd.read_csv(url)

# Criando a regra de classificaГ§ГЈo com base na mediana (Dados 100% equilibrados)
preco_medio = dados['price'].median()
dados['bom_negocio'] = np.where(dados['price'] <= preco_medio, 1, 0)

X = dados[['year', 'mileage']]
y = dados['bom_negocio']

print(f" -> Dataset 'usedcars.csv' carregado com sucesso!")
print(f" -> Total de veГ­culos para anГЎlise: {len(dados)}")
print(f" -> PreГ§o mediano de corte: ${preco_medio:.2f}")
print(f" -> Objetivo: Classificar se o carro Г© um 'Bom NegГіcio' (PreГ§o <= Mediana)")

# =====================================================================
# ETAPA 2: ANГЃLISE DOS DADOS E PRГ‰-PROCESSAMENTO
# =====================================================================
print("\n" + "="*60)
print(" ETAPA 2: ANГЃLISE E PRГ‰-PROCESSAMENTO ".center(60, "="))
print("="*60)

# Redes Neurais exigem que os dados numГ©ricos estejam na mesma escala (padronizados)
scaler = StandardScaler()
X_escalonado = scaler.fit_transform(X)

# DivisГЈo dos dados: 70% para treino e 30% para teste
X_treino, X_teste, y_treino, y_teste = train_test_split(X_escalonado, y, test_size=0.3, random_state=42)

print(f" [OK] Recursos ('year' e 'mileage') padronizados com StandardScaler.")
print(f" -> Amostras de Treino: {X_treino.shape[0]} carros")
print(f" -> Amostras de Teste: {X_teste.shape[0]} carros")

# =====================================================================
# ETAPA 3: TREINAMENTO (REDES NEURAIS ARTIFICIAIS)
# =====================================================================
print("\n" + "="*60)
print(" ETAPA 3: TREINAMENTO DA REDE NEURAL (KERAS) ".center(60, "="))
print("="*60)

# Construindo a Rede Neural com o Keras (Sequential)
modelo_neural = Sequential()

# Camada de Entrada + 1ВЄ Camada Oculta (com 12 neurГґnios e ativaГ§ГЈo ReLU)
modelo_neural.add(Dense(units=12, activation='relu', input_shape=(X_treino.shape[1],)))

# 2ВЄ Camada Oculta (Atendendo a exigГЄncia do enunciado: pelo menos 2 camadas ocultas!)
modelo_neural.add(Dense(units=8, activation='relu'))

# Camada de SaГ­da (1 neurГґnio com Sigmoid para classificaГ§ГЈo binГЎria: 0 ou 1)
modelo_neural.add(Dense(units=1, activation='sigmoid'))

# Compilando a Rede Neural com os parГўmetros para classificaГ§ГЈo binГЎria
modelo_neural.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

print(" -> Iniciando o treinamento da Rede Neural (Backpropagation)...")
print("-" * 60)
# O modelo treina passando pelos dados 30 vezes (epochs)
historico = modelo_neural.fit(X_treino, y_treino, epochs=30, batch_size=5, validation_split=0.1, verbose=1)
print("-" * 60)
print(" [SUCESSO] Rede Neural treinada com ГЄxito!")

# =====================================================================
# ETAPA 4: VALIDAГ‡ГѓO E MГ‰TRICAS
# =====================================================================
print("\n" + "="*60)
print(" ETAPA 4: VALIDAГ‡ГѓO E MГ‰TRICAS DE DESEMPENHO ".center(60, "="))
print("="*60)

# Fazendo as previsГµes com o lote de teste
previsoes_probabilidade = modelo_neural.predict(X_teste)
previsoes = np.where(previsoes_probabilidade >= 0.5, 1, 0)

# Calculando a acurГЎcia (taxa de acerto de 0 a 100%)
acuracia_final = accuracy_score(y_teste, previsoes)
print(f" -> AcurГЎcia obtida na base de Teste: {acuracia_final * 100:.2f}%")

print("\n" + "-"*55)
print(" RELATГ“RIO DE CLASSIFICAГ‡ГѓO DETALHADO ".center(55, " "))
print("-"*55)
print(classification_report(y_teste, previsoes))
print("-"*55 + "\n")
```
